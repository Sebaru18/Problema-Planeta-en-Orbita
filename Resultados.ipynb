a continuación se mostrara los resultados, estos se mostraran en una tabla, se traera el archivo csv que se genero al ejecutar el performance.py y este csv se llamara tierra.csv

```python
import pandas as pd
resultados = pd.read_csv('/content/tierra.csv')
resultados.head()
```

Para poder entender la tabla se le colocara los nombres a cada columna, pues la columna de la izquierda es la columna de la ejecución en Python, mientras que el de la derecha es el de la ejecucion de Cython.

```python
resultados.columns=['Python','Cython']
resultados
```

```python
print(" El promedio de Python es de: ", resultados['Python'].mean())
```

```python
print(" El promedio de Cython es de: ", resultados['Cython'].mean())
```

Como se puede observar la diferencia de rendimiento es notoria, ya que se evidencia que Cython es mas eficaz al ejecutar el programa, teniendo un promedio de 0.346s, mientras que el promedio de la ejecución en Python fue de 6.039s, siendo la diferencia aproximadamente de 5.593s

Ahora, la ejecución y la toma de datos en las pruebas de rendimiento se realizaron con las siguientes caracteristicas.
